# Import Packages

```python
#Import Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import Relevant CSV Files
pop = pd.read_csv('C:/Users/zalle/Documents/GitHub/capstone-1/data/Originals/population_stats.csv')
e_deaths = pd.read_csv('C:/Users/zalle/Documents/GitHub/capstone-1/data/Originals/estimated_deaths.csv')
m_deaths = pd.read_csv('C:/Users/zalle/Documents/GitHub/capstone-1/data/Originals/reported_deaths.csv')
resistance = pd.read_csv('C:/Users/zalle/Documents/GitHub/capstone-1/data/Originals/insecticide_resistance.csv')
net_dist = pd.read_csv('C:/Users/zalle/Documents/GitHub/capstone-1/data/Originals/amf_distributions4.csv')                         
```

```python
#Display the original population data for reference
print(pop.head())
```

```python
pop = pop.rename(columns={'Display Value' : 'country_pop'})
pop = pop.rename(columns={'YEAR (CODE)' : 'year'})
pop = pop.rename(columns={'COUNTRY (CODE)' : 'country_code'})
```

```python
#Clear irrelevant columns from population data
pop_revised = pop[['year', 'REGION (CODE)', 'country_code', 'COUNTRY (DISPLAY)', 'country_pop']]
print(pop_revised.head())
```

## Estimated Deaths File

```python
print(e_deaths.head())
```

```python
e_deaths = e_deaths.rename(columns={'Numeric':'estimated_death_Numeric', 'Low':'estimated_death_Low', 'High':'estimated_death_High'})
e_deaths = e_deaths.rename(columns={'YEAR (CODE)' : 'year'})
e_deaths = e_deaths.rename(columns={'COUNTRY (CODE)' : 'country_code'})
```

```python
# Revised yearly reported deaths by country to be merged
e_deaths_revised = e_deaths[['year', 'country_code', 'estimated_death_Numeric', 'estimated_death_Low', 'estimated_death_High']]
print(e_deaths_revised.head())
```

## Measured Deaths File

```python
print(m_deaths.head())
```

```python
m_deaths = m_deaths.rename(columns={'Numeric':'measured_death_Numeric'})
m_deaths = m_deaths.rename(columns={'YEAR (CODE)' : 'year'})
m_deaths = m_deaths.rename(columns={'COUNTRY (CODE)' : 'country_code'})
```

```python
# Revised yearly measured deaths by country to be merged
m_deaths_revised = m_deaths[['year', 'REGION (CODE)', 'country_code', 'measured_death_Numeric']]

print(m_deaths_revised.head())
```

## Resistance File

```python
print(resistance.head())
```

```python
resistance = resistance.rename(columns={'Display Value':'tx_resistance'})
resistance = resistance.rename(columns={'YEAR (CODE)' : 'year'})
resistance = resistance.rename(columns={'COUNTRY (CODE)' : 'country_code'})
```

```python
# Standardizing responses in the resistance column: '0' for 'not tested/reported' and 'no' responses, '1' for Yes i.e. countries with known resistance
resistance_to_int = {'Yes':1,
                      'No':0,
                      'Not tested/reported':0,
                      np.nan:0}
resistance['tx_resistance_int'] = resistance['tx_resistance'].map(resistance_to_int)
print('# variables in column:', resistance['tx_resistance_int'].nunique())
```

```python
# Revised countries with reported resistance
resistance_revised = resistance[['year', 'country_code', 'tx_resistance_int']]
print(resistance_revised.head())
```

## Net Distributions File

```python
# Date ranges need to be made consistant for comparison with other data
print(net_dist.head())
net_dist.info()
```

```python
net_dist['number_llins'].head()
```

```python
# Parse out year and add year code column to the 'net_dist'        
net_dist['YEAR (CODE)'] = '20'+net_dist['when'].str[-2:]
net_year = pd.to_numeric(net_dist['YEAR (CODE)'], errors='coerce')
net_dist['year'] = net_year
net_dist.info()
print(net_dist)
```

```python
print('Year range for net distribution data:','',net_dist['year'].nunique())
print('Number of countries:','',net_dist['country_code'].nunique())
```

```python
# Net df revised
net_dist_revised = net_dist[['year', 'country_code', 'number_llins', 'by_whom']]
```

## Merge with m_deaths_revised as base

```python
df1 = pd.merge(m_deaths_revised, e_deaths_revised, how='left', on=['year','country_code'])
df2 = df1.merge(resistance_revised, how='left', on=['year','country_code'])
```

```python
df_working = df2.merge(net_dist_revised, how='left', on=['year', 'country_code'])
df_working.info()
print(df_working.head())
```

```python
df_working.info()
print(df_working.columns)
print(df_working.describe())
print(df_working.shape)
print('Number of years:', '', df_working['year'].nunique())
print('Years from:', '', df_working['year'].min(), '-', df_working['year'].max())
print('Number of countries:', '', df_working['country_code'].nunique())
```
