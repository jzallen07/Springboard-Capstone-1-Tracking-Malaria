```python
import matplotlib.pyplot as plt
import matplotlib
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
import seaborn as sns 
import numpy as np
import pandas as pd
import numpy as np
import random as rnd
import re
import io

%matplotlib inline
```

```python
cases = pd.read_csv('data/Originals/reported_indigenous_confirmed_cases.csv')
```

```python
# renaming columns
cases = cases.rename(columns={'YEAR (CODE)':'year',
                    'REGION (CODE)':'region',
                    'COUNTRY (CODE)':'country_code',
                    'REGION (CODE)' : 'region',
                     'COUNTRY (DISPLAY)' : 'country_name',         
                    'Numeric':'reported_cases'})

# selecting relevant columns
cases = cases[['year', 'country_name','reported_cases', 'region']]

# examining the df
print('Years from:', '', cases['year'].min(), '-', cases['year'].max())
print(cases.isnull().any())
print(cases.head())
print(cases.info())
```

```python
cases_sum = cases.groupby(['year', 'country_name']).sum()
```

```python
cases_sum.head()
```

```python
cases_sum = cases_sum.reset_index()
```

```python
cases_sum.columns
```

```python
cases_2000 = cases_sum[(cases_sum['year'] == 2000)]
```

```python
cases_2000
```

```python
data =  dict(
        type = 'choropleth',
        locations = cases_2000.country_name,
        locationmode = 'country names',
        z = cases_2000.reported_cases,
        text = cases_2000.country_name,
        autocolorscale = False,
        colorscale = [[0, 'rgb(255,237,160)'],[0.5, 'rgb(254,178,76)'],[1, 'rgb(240,59,32)']],
        colorbar = {'title': 'No. of Cases'})

layout = dict( title = 'Reported Cases of Malaria 2000',
         geo = dict(scope = 'world',
                    showframe = True,
                    showrivers = True,
                    showlakes = True,
                    showcountries = True,
                    showocean = True,
                    oceancolor = '#deebf7',
                    resolution = '110', 
         projection = {'type' : 'equirectangular'}))

choromap3 = go.Figure(data = [data],layout=layout)
iplot(choromap3)
```

```python
cases_2014 = cases_sum[(cases_sum['year'] == 2014)]
```

```python
data =  dict(
        type = 'choropleth',
        locations = cases_2014.country_name,
        locationmode = 'country names',
        z = cases_2014.reported_cases,
        text = cases_2014.country_name,
        autocolorscale = False,
        colorscale = [[0, 'rgb(255,237,160)'],[0.5, 'rgb(254,178,76)'],
                      [1, 'rgb(240,59,32)']],
        colorbar = {'title': 'No. of Cases'})

layout = dict( title = 'Reported Cases of Malaria 2014',
         geo = dict(scope = 'world',
                    showframe = True,
                    showrivers = True,
                    showlakes = True,
                    showcountries = True,
                    countrycolor = 'rgb(99,99,99)',
                    showocean = True,
                    oceancolor = '#deebf7',
                    resolution = '110', 
         projection = {'type' : 'equirectangular'}))

choromap3 = go.Figure(data = [data],layout=layout)
iplot(choromap3)
```
